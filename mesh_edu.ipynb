```python
import numpy as np
import pandas as pd
import nltk
```

```python
from nltk import tokenize
from sklearn.feature_extraction.text import CountVectorizer
from nltk.stem.porter import PorterStemmer
from nltk.stem.lancaster import LancasterStemmer
from nltk.corpus import stopwords
import re
from sklearn.feature_extraction.text import TfidfTransformer,TfidfVectorizer
```

```python
txt=open("G:/nlp_meshedu/training.txt","r")
```

```python
txt=txt.read()
newlinesep = txt.split('\n')
newlinesep = newlinesep[1:]
```

```python
sents=[]
target=[]
for i in newlinesep:
    if len(i.split('\t'))>1:
        sents.append(i.split('\t')[0])
        target.append(i.split('\t')[1])
    else:
        print(i)
```

```python
for no,i in enumerate(sents):
    sents[no]=re.sub('[^a-zA-z\s]',' ',sents[no])
    sents[no]=sents[no].lower()
    target[no]=re.sub('[^a-zA-z\s]',' ',target[no])
    target[no]=target[no].lower()
```

```python
target=np.array(target)
sents=np.array(sents)
lstem = LancasterStemmer()
for no,i in enumerate(sents):
    sents[no]=" ".join([lstem.stem(j) for j in tokenize.word_tokenize(i)])
#adding a suffix after one lettered words..
for no,i in enumerate(sents):
    sents[no]=" ".join([j+"suffix" if len(j)==1 else j for j in i.split(" ")])
for no,i in enumerate(target):
    target[no]=" ".join([lstem.stem(j) for j in tokenize.word_tokenize(i)])
#adding a suffix after one lettered words..
for no,i in enumerate(target):
    target[no]=" ".join([j+"suffix" if len(j)==1 else j for j in i.split(" ")])
```

```python
cnt1=TfidfVectorizer(ngram_range=(1,2),stop_words="english")
cnt1.fit(np.concatenate((target,sents)))
arr1=cnt1.transform(sents)
print(arr1.shape)
X=arr1.toarray()
```

```python
arr2=cnt1.transform(target)
print(arr2.shape)
y=arr2.toarray()
```

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC
from sklearn.gaussian_process import GaussianProcessClassifier
from sklearn.gaussian_process.kernels import RBF
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier, AdaBoostClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.discriminant_analysis import QuadraticDiscriminantAnalysis
from sklearn.pipeline import Pipeline
from sklearn.linear_model import SGDClassifier
from sklearn.svm import LinearSVC
```

```python
names = ["Nearest Neighbors", "Linear SVM", "RBF SVM", "Gaussian Process",
         "Decision Tree", "Random Forest", "AdaBoost",
         "Naive Bayes", "QDA"]

classifiers = [
    SVC(kernel="linear", C=0.025),
    SVC(gamma=2, C=1),
    DecisionTreeClassifier(max_depth=5),
    RandomForestClassifier(max_depth=5, n_estimators=10, max_features=1),
]
```

```python
for clf in classifiers:
        clf.fit(X, target)
        print(clf.score(X, target))
```

# here the svc with rbf kernel performs well...0.99 accuracy i trained it on whole dataset as i wanted it to overfit it as the feature matrix is gonna be very sparse..

```python
text_clf=Pipeline([('clf', SVC(gamma=2, C=1))])
text_clf.fit(X,target)
```

```python
test=[]
test_sen=[]
for i in range(int(input())): 
    x=input()
    x=re.sub('[^a-zA-z\s]',' ',x)
    x=x.lower()
    x=" ".join([lstem.stem(j) for j in tokenize.word_tokenize(x)])
    x=" ".join([j+"suffix" if len(j)==1 else j for j in x.split(" ")])
    sen=" ".join(word for word in tokenize.word_tokenize(x))
    trans_sen=cnt1.transform(np.array([sen]))
    test.append(trans_sen.toarray()[0])
    test_sen.append(sen)
predicted=text_clf.predict(np.array(test))
for i,j in zip(predicted,test_sen):
    print(j +" ::: this is the query most likely used... \n"+ i)
```
